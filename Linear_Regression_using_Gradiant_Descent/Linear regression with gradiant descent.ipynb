#Gradient Descent
1. You start with points, starting inter and slope value, Learning rate, No. of iterations

2. For number of iterations
    
    2.1 Calculate gradiant (Kis direction me sudharna hai slope and inter ko)
    
    2.2 Kitna se sudharna hai (Learning rate)
    
    2.3 2.1 and 2.2 mila ke new slope and inter

3. Last me compute error

- Pehle intercept zero man ke error nikatlte hai 

```python
def error(x,y,slope):
    num_columns = len(x)
    tmp =0
    for i in range(num_columns):
        tmp+=(y[i] - (slope * x[i])) ** 2
    return (tmp/float(num_columns))
    
def gradient_descent_runner(x,y,slope):
    num_columns = len(x)
    tmp =0
    for i in range(num_columns):
        tmp+=((-1 * x[i]) * (y[i] - (slope * x[i])))
    return tmp * float(2 / num_columns)

def gradient_descent(x,y,slope, learning_rate):
    grad = gradient_descent_runner(x,y,slope)
    print(f"Gradiant: {grad}")
    new_slope = slope - (learning_rate * grad)
    return new_slope

def init(x,y):
    slope = 0
    num_iteration = 1000
    learning_rate = 0.0001
    slope_list = []
    error_list = []
    
    for i in range(num_iteration):
        slope = gradient_descent(x,y,slope, learning_rate)
        slope_list.append(slope)
        
        err = error(x,y,slope)
        error_list.append(err)
        print(f"Iteration: {i} Slope: {slope} Error: {err}")
    
    plt.plot(slope_list, error_list)
    plt.xlabel("Slope")
    plt.ylabel("Error")
    plt.figure(figsize=(10,8))
    
    print(f"New slope: {slope}")
    print(f"Error: {error(x,y,slope)}")

init(data.Population, data.Profit)
```

```python
def error(x,y,slope, inter):
    num_columns = len(x)
    tmp =0
    for i in range(num_columns):
        tmp+=(y[i] - (slope * x[i] + inter)) ** 2
    return (tmp/float(num_columns))
    
def gradient_descent_runner(x,y,slope,inter):
    num_columns = len(x)
    tmp_slope =0
    tmp_inter = 0
    for i in range(num_columns):
        tmp_slope+=((-1 * x[i]) * (y[i] - (slope * x[i] + inter))) * float(2 / num_columns)
        tmp_inter+=(-1 * (y[i] - (slope * x[i] + inter)))* float(2 / num_columns)
    return tmp_slope, tmp_inter

def gradient_descent(x,y,slope, inter, learning_rate):
    grad_slope, grad_inter = gradient_descent_runner(x,y,slope, inter)
    print(f"Gradiant Slope: {grad_slope} Inter: {grad_inter}")
    new_slope = slope - (learning_rate * grad_slope)
    new_inter = inter - (learning_rate * grad_inter)
    return new_slope, new_inter

def init(x,y):
    slope = 0
    inter = 0
    num_iteration = 1000
    learning_rate = 0.0001
    
    for i in range(num_iteration):
        slope, inter = gradient_descent(x,y,slope, inter, learning_rate)        
        err = error(x,y,slope, inter)
        print(f"Iteration: {i} Slope: {slope} Inter: {inter} Error: {err}")
    
    print(f"New slope: {slope}")
    print(f"New inter: {inter}")
    print(f"Error: {error(x,y,slope, inter)}")

init(data.Population, data.Profit)
```

```python
def return_y_wrt_x(x,slope,intercept=0):
    return ((slope * x) + intercept)
```
